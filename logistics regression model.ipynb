### read the file and vectorization

```python
import pandas as pd
df=pd.read_csv("yelp_cleaned_data.csv")
```

```python
import numpy as np
```

```python
good=df[df['class']==1]
bad=df[df['class']==0]
```

```python
df1=df[['text','class']]
from nltk.corpus import stopwords
wd=stopwords.words()+['worse','yeah','yes','no','absolutely','actual','add','ago',
                     'actually','able','yet','wrong','hands','disappointed','disappointing',
                      'bad','good','dont','better','bacchanal','worst','poor','buffet','food',
                      'good','like','best','worth','vegas','back','tried','year','story','definitely',
                      'really','delicious','favourite','great','well','work','ago','added','additional','would','will',
                     'along','alone','almost','without','wonderful','world','wow','although','yet','wish','within','hope',
                     'afternoon','ahead','already','always','amazing','within','alright','anything','amount','anyone','willing',
                     'another','anything','however','but','end','anymore','anyway','anywhere','disappointment','someone',
                     'sick','horrible','penny','means','sometimes','charge','means','front','asked','array','arrive',
                     'around','area','areas','yelp','working','yum','wife','whatever','arrived','wide','years','yummy','whole',
                     'word','went','go','ask','ate','weird','wise','way']
from sklearn.feature_extraction.text import CountVectorizer
vectorizer=CountVectorizer(max_features=1000,stop_words= wd,token_pattern='(?u)\\b[a-zA-z][a-zA-Z]+\\b') 
```

```python
X=vectorizer.fit_transform(df1['text'])
y=df1['class'].values
```

```python
df=pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names())
df.head()
```

### Build logitics regression model

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
result=lr.fit(X,y)
```

```python
y_pred=lr.predict(X)
np.mean(y_pred==y)   ### accuracy
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y,y_pred)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y,y_pred)   ### AUC
```

### calculate feature importance

```python
feature_importance = abs(lr.coef_[0])
```

```python
feature_importance = 100.0 * (feature_importance / feature_importance.max())
```

```python
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5
```
