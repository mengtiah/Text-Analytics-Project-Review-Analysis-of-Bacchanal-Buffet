```python
import pandas as pd
df=pd.read_csv('yelp_cleaned_data.csv')
```

```python
## devide our reviews to good and bad
good=df[df['class']==1]
bad=df[df['class']==0]
```

```python
good_reviews=list(good['text'].values)
bad_reviews=list(bad['text'].values)
```

# TF-IDF for good reviews

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
wd=stopwords.words()+['bacchanal','buffet','food','good','like','best','worth','vegas','back',
                      'tried','year','story','definitely','really',"get", "ever","take", 
                      "eat", "go", "said",'buffets','highly','recommend','king']

vectorizer = TfidfVectorizer(ngram_range=(3,6),
                             token_pattern=r'\b[a-zA-Z]{3,}\b',
                             max_df=0.4,
                             min_df=2, stop_words=wd)
```

```python
from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize
import string
lemmatizer = WordNetLemmatizer()
def lemmatize_sentence(sentence):
    words = word_tokenize(sentence)
    res_words = []
    for word in words:
        res_words.append(lemmatizer.lemmatize(word).strip(string.punctuation))
    return " ".join(res_words)
```

```python
X = vectorizer.fit_transform(good_reviews)
terms = vectorizer.get_feature_names()
tf_idf= pd.DataFrame(X.toarray().transpose(), index=terms)
tf_idf = tf_idf.sum(axis=1)
score = pd.DataFrame(tf_idf, columns=["score"])
score["term"] = terms
score.sort_values(by="score", ascending=False, inplace=True)
```

# TF-IDF for bad Reviews

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
wd=stopwords.words()+['buffet','crab','worth','worst','back','eating','eat','good','forward',
                      'looking','better','great','best','story','waste']

vectorizer = TfidfVectorizer(ngram_range=(3,6),
                             token_pattern=r'\b[a-zA-Z]{3,}\b',
                             max_df=0.4,
                             min_df=2, stop_words=wd)
```

```python
Y = vectorizer.fit_transform(bad_reviews)
terms = vectorizer.get_feature_names()
tf_idf1= pd.DataFrame(Y.toarray().transpose(), index=terms)
tf_idf1 = tf_idf1.sum(axis=1)
score1 = pd.DataFrame(tf_idf1, columns=["score"])
score1["term"] = terms
score1.sort_values(by="score", ascending=False, inplace=True)
```
